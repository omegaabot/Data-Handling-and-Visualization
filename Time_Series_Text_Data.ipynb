```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Date": pd.to_datetime(["2020-01-01", "2021-02-01", "2022-03-01", "2023-04-01", "2024-01-01"]),
    "Value": [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker="X")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Month": ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"],
    "Sales": [100, 150, 130, 162, 180, 200, 225, 205, 195, 220, 225, 230]
}

df = pd.DataFrame(data)
df.set_index("Month", inplace=True)

plt.figure(figsize=(10, 6))
plt.plot(df.index, df["Sales"], marker="o")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=69)
plt.title("Time Series with Seasonality")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
text = "This is a sample text for creating a word cloud. It contains words with different frequencies."
wordcloud = WordCloud().generate(text)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

words=text.split()
word_counts=Counter(words)
plt.figure(figsize=(10,5))
plt.xticks(rotation=69)
plt.bar(word_counts.keys(),word_counts.values())
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def remove_stop_words(string):
  stop_words=set(stopwords.words('english'))
  words=string.split()
  filtered_words=[word for word in words if word.lower() not in stop_words]
  new_string=' '.join(filtered_words)
  return new_string
input_string="This is an example sentence to rename stop words"
result=remove_stop_words(input_string)
print("Original string",input_string)
print("Modified string",result)
```
