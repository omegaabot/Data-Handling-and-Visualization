```python
# Import the necessary modules
import folium

# Create a map
mymap = folium.Map(location=[13.05, 77.6], zoom_start=12)  # Adjust location and zoom level as needed

# Create a list of locations with coordinates and image URLs
locations = [
    {"name": "Electronic City", "lat": 12.9985954, "lon": 77.5920086, "img": "https://im.hunt.in/cg/City-Guide/elec.jpg", "bounds": [ [12.98, 77.58], [13.02, 77.60] ]},
    {"name": "Bannerghatta Road", "lat": 12.9596466, "lon": 77.5736842, "img": "https://static.toiimg.com/photo/40193882/.jpg", "bounds": [ [12.94, 77.56], [12.98, 77.58] ]},
    {"name": "HSR Layout", "lat": 12.9797454, "lon": 77.5906167, "img": "https://upload.wikimedia.org/wikipedia/commons/7/7b/2%2C_Park_-_HSR_Layout%2C_Bengaluru.jpg", "bounds": [ [12.97, 77.58], [12.98, 77.59] ]},
    {"name": "Koramangala", "lat": 12.974244, "lon": 77.592195, "img": "https://imgmediagumlet.lbb.in/media/2022/01/61efa39514a08c2400a7ce56_1643094933277.png?fm=webp&w=750&h=500&dpr=1", "bounds": [ [12.97, 77.59], [12.98, 77.60] ]}
]

# Add markers with custom popups
for location in locations:
    # Create a marker for the location
    marker = folium.Marker(location=[location["lat"], location["lon"]])

    # Create a custom popup with the image and location name
    popup_html = folium.Popup(folium.IFrame(html=f'<b>{location["name"]}</b><br><img src="{location["img"]}" width="200">', width=200, height=200))

    # Add the popup to the marker
    marker.add_child(popup_html)

    # Add the marker to the map
    marker.add_to(mymap)

# Display the map
mymap
```
